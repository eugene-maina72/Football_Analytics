##Football Pitch Layout

```python
import matplotlib.pyplot as plt
from mplsoccer import Pitch, VerticalPitch 
```

```python
#Ploting a paitch on a new axis
pitch=Pitch()
#Specify the size of the pitch
fig, ax = pitch.draw(figsize=(8, 4))
```

```python
fig, axs = plt.subplots(nrows=1, ncols=2)
pitch =Pitch()
pie = axs[0].pie(x=[5, 15])
pitch.draw(ax=axs[1])
```

```python
#Drawing pitch templates they include (Opta ,Wyscout,StatsBomb,Tracab,Uefa,metricasports,custom,skillcorner,secondspectrum and impect)
#For (‘metricasports’, ‘tracab’, ‘skillcorner’, ‘secondspectrum’ or ‘custom’) you also need to specify the pitch_length and pitch_width, which are typically 105 and 68 respectively.
pitch =Pitch(pitch_type='opta')
fig, ax = pitch.draw()
```

```python
pitch = Pitch(pitch_type='skillcorner',
              pitch_length=105, pitch_width=68
              , axis=True, label=True)
fig, ax = pitch.draw()  
```

```python
pitch = Pitch()
fig, axs = pitch.draw(nrows=2, ncols=3)
```

```python
pitch = Pitch()
fig, axs = pitch.draw(nrows=2, ncols=3, tight_layout=True, constrained_layout=True)
```

```python
pitch = Pitch()
fig, axs = pitch.grid(nrows=2, ncols=3, figheight=10,
                      #Grid takes up 71.5% of the figure height
                      grid_height=0.715,
                      #5% of the figure height is reserved for space inbetween the axes
                      space=0.05,
                      #Centres the grid horizontally and vertically
                      left=None, bottom=None
                      )
```

```python
pitch = Pitch(half= False)
fig, ax = pitch.draw()  
```

```python
pitch = VerticalPitch(half=True)
fig, ax = pitch.draw()
```

```python
pitch = VerticalPitch(half=True,
                      pad_left=-10,  # bring the left axis in 10 data units (reduce the size)
                      pad_right=-10,  # bring the right axis in 10 data units (reduce the size)
                      pad_top=10,  # extend the top axis 10 data units
                      pad_bottom=20)  # extend the bottom axis 20 data units
fig, ax = pitch.draw()
```

```python
pitch=Pitch(pitch_color='#aabb97', line_color='white', stripe_color='#c2d59d', stripe=True, linestyle='--', linewidth=1, goal_linestyle='-', corner_arcs=True)
fig, ax = pitch.draw()
```

```python
pitch = Pitch(line_alpha=0.5, goal_alpha=0.3)
fig, ax = pitch.draw()
```
